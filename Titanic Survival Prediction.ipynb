```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean()).astype(int)
titanic_df['Fare'] = titanic_df['Fare'].fillna(titanic_df['Fare'].mean()).astype(int)
```

```python
titanic_df['Embarked'].unique()
```

```python
titanic_df['Embarked'] = titanic_df['Embarked'].map( {'Q': 0,'S':1,'C':2}).astype(int)
titanic_df['Sex'] = titanic_df['Sex'].map( {'female': 1,'male':0}).astype(int)
```

```python
titanic_df.dtypes
```

```python
data = titanic_df.drop(['PassengerId','Name','Cabin','Ticket'], axis =1, inplace=True)
```

# spliting the data

```python
Train = titanic_df.drop(['Survived'], axis=1)
Test = titanic_df.iloc[:,1]
x_train, x_test, y_train, y_test = train_test_split(Train, Test, test_size = 0.2, random_state = 1)
```

# Model Building

```python
LR = LogisticRegression(solver='liblinear', max_iter=200)
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
LRAcc = accuracy_score(y_pred,y_test)
print('Logistic regression accuracy: {:.2f}%'.format(LRAcc*100))
```
